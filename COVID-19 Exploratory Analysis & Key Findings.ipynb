# Importing Libraries & Loading Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load the dataset

```python
data=pd.read_csv("country_wise_latest.csv")
```

# Display basic info and first few rows

```python
data.info()
data.head()
```

# Data Overview & Summary Statistics

```python
#Check for midding values
print(data.isnull().sum())
```

```python
data.describe()
```

# Top 10 Most Affected Countries

```python
# Set seaborn theme
sns.set(style="whitegrid")

# Sort and select top 10 countries by confirmed cases
top_10 = data.sort_values(by='Confirmed', ascending=False).head(10)

# Create figure and plot
plt.figure(figsize=(12, 5))
barplot = sns.barplot(
    x='Confirmed',
    y='Country/Region',
    data=top_10,
    palette='Reds_r'  # Reversed Reds color palette
)

# Add title and axis labels
plt.title('Top 10 Most Affected Countries', fontsize=16, fontweight='bold')
plt.xlabel('Confirmed Cases', fontsize=12)
plt.ylabel('Country', fontsize=12)

# Add value labels to each bar
for index, value in enumerate(top_10['Confirmed']):
    plt.text(value + 5000, index, f'{int(value):,}', va='center', fontsize=10)

# Adjust layout
plt.tight_layout()
plt.show()

```

```python
# Select only numeric columns
numeric_df = data.select_dtypes(include=['number'])

# Compute correlation matrix
corr_matrix = numeric_df.corr()

# Set dark theme
sns.set_theme(style="darkgrid")

# Create the figure
plt.figure(figsize=(14,8))

# Draw the heatmap with enhanced style
heatmap = sns.heatmap(
    corr_matrix,
    annot=True,
    fmt=".2f",
    cmap='coolwarm',         # Gradient from cool to warm
    center=0,
    square=True,
    linewidths=0.7,
    linecolor='black',
    cbar_kws={
        "shrink": 0.9,
        "aspect": 30,
        "pad": 0.02,
        "orientation": "vertical"
    }
)

# Add a title with styling
plt.title("Correlation Heatmap of COVID-19 Metrics", fontsize=20, fontweight='bold', color='white', pad=30)

# Customize tick labels
plt.xticks(rotation=45, ha='right', fontsize=12, color='white')
plt.yticks(rotation=0, fontsize=12, color='white')

# Customize colorbar
colorbar = heatmap.collections[0].colorbar
colorbar.ax.tick_params(labelsize=10, colors='white')  # Colorbar tick labels in white
colorbar.set_label('Correlation', color='white', fontsize=12)

# Set dark background
plt.gca().set_facecolor('#1e1e1e')        # Axis background
plt.gcf().patch.set_facecolor('#1e1e1e')  # Figure background

# Tight layout for better spacing
plt.tight_layout()

# Show the heatmap
plt.show()
```

# Death Rate vs. Recovery Rate

```python
# Scatter plot: Death Rate vs. Recovery Rate
plt.figure(figsize=(12, 4))
sns.scatterplot(x=data['Deaths / 100 Cases'], y=data['Recovered / 100 Cases'], alpha=0.7)
plt.xlabel("Death Rate (%)")
plt.ylabel("Recovery Rate (%)")
plt.title("Death Rate vs. Recovery Rate")
plt.grid(True)
plt.show()
```

# Weekly % Increase in Cases

```python
# Top 10 countries with the highest weekly % increase
top_growth2 = data[['Country/Region', '1 week % increase']].sort_values(by='1 week % increase', ascending=False).head(10)

# Bar plot
plt.figure(figsize=(10, 4))
sns.barplot(x='1 week % increase', y='Country/Region', data=top_growth2, palette='Blues_r')
plt.xlabel("1 Week % Increase")
plt.ylabel("Country")
plt.title("Top 10 Countries with Highest Weekly Growth")
plt.show()
```

# Active vs. Recovered Cases by Country

```python
#Scatter plot: Active vs Recovered Cases by country:-
plt.figure(figsize=(10,5))
sns.scatterplot(x=data["Active"], y=data["Recovered"], alpha=0.7)
plt.xlabel("Active Cases")
plt.ylabel("Recovered Cases")
plt.title("Active vs Recovered Cases by country.")
plt.grid(True)
plt.show()
```

# Mortality Rate Across WHO Regions

```python
# Boxplot of death rate by WHO regio
plt.figure(figsize=(10, 4))
sns.boxplot(x=data['WHO Region'], y=data['Deaths / 100 Cases'], palette='coolwarm')
plt.xlabel("WHO Region")
plt.ylabel("Death Rate (%)")
plt.title("Mortality Rate Across WHO Regions")
plt.xticks(rotation=45)
plt.show()
```

# Recovery Rate Distribution

```python
# Histogram of recovery rate
plt.figure(figsize=(8, 4))
sns.histplot(data['Recovered / 100 Cases'], bins=20, kde=True, color='blue')
plt.xlabel("Recovery Rate (%)")
plt.ylabel("Frequency")
plt.title("Distribution of Recovery Rates Across Countries")
plt.show()
```

# Countries with Zero Active Cases

```python
# Countries with zero active cases
zero_active = data[data['Active'] == 0]

# Count plot
plt.figure(figsize=(10, 3))
sns.countplot(y=zero_active['WHO Region'], palette='Greens_r')
plt.xlabel("Count")
plt.ylabel("WHO Region")
plt.title("Countries with Zero Active Cases by WHO Region")
plt.show()
```
